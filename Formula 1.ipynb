```python
import os
import pandas as pd
import numpy as np
import fastf1
fastf1.logger.set_log_level('INFO')
# Create cache once
os.makedirs("f1_cache", exist_ok=True)
fastf1.Cache.enable_cache("f1_cache")
def load_clean_laps(year: int, event: str, session_code: str = 'R'):
    """
    Returns a cleaned lap-level DataFrame for a race with core features engineered.
    """
    ses = fastf1.get_session(year, event, session_code)
    ses.load()  # needs internet on first run; then uses cache

    laps = ses.laps.copy()

    # --- Basic cleaning
    # Exclude inaccurate or deleted laps
    laps = laps[(laps["IsAccurate"] == True) & (laps["Deleted"] == False)]

    # Exclude in-laps and out-laps
    laps = laps[laps["PitInTime"].isna() & laps["PitOutTime"].isna()]

    # Exclude SC/VSC/Yellow laps -> keep 'green' only.
    # TrackStatus can be NaN (treat as green) or codes like '1' (= green).
    def is_green(x):
        if pd.isna(x): 
            return True
        # Some laps have multiple codes like '1;2'; keep strictly green-only laps
        return str(x).strip() == '1'
    laps = laps[laps["TrackStatus"].apply(is_green)]

    # Drop wet compounds initially (start with dry model; add later)
    laps = laps[~laps["Compound"].isin(["INTERMEDIATE", "WET"])]

    # --- Feature engineering
    # Timedelta -> seconds
    laps["LapTime_s"] = laps["LapTime"].dt.total_seconds()

    # Tyre age (already there as TyreLife, but ensure numeric)
    laps["TyreAge"] = laps["TyreLife"].astype(float)

    # Stint-relative lap index (helps model degradation within stint)
    laps["StintLap"] = laps.groupby(["Driver", "Stint"])["LapNumber"] \
                           .rank(method="first").astype(int)

    # Driver pace normalization options (pick later in modeling)
    # 1) Delta to each driver's race median
    laps["DeltaToDriverMedian_s"] = laps["LapTime_s"] - laps.groupby("Driver")["LapTime_s"].transform("median")
    # 2) Delta to each driver's best (more aggressive)
    laps["DeltaToDriverBest_s"] = laps["LapTime_s"] - laps.groupby("Driver")["LapTime_s"].transform("min")

    # Rolling form (last 3 laps average) within stint
    laps["StintRoll3_s"] = laps.sort_values(["Driver","Stint","LapNumber"]) \
        .groupby(["Driver","Stint"])["LapTime_s"].transform(lambda s: s.rolling(3, min_periods=1).mean())

    # One-hot encode compound (or leave as category for tree models)
    laps["Compound"] = laps["Compound"].astype("category")

    # Merge weather (nearest time to lap start)
    wthr = ses.weather_data  # columns: Time, AirTemp, Humidity, Pressure, Rainfall, TrackTemp, WindDirection, WindSpeed
    # align by nearest lap start time
    laps = pd.merge_asof(
        laps.sort_values("LapStartTime"),
        wthr.sort_values("Time").rename(columns={"Time":"WeatherTime"}),
        left_on="LapStartTime", right_on="WeatherTime", direction="nearest"
    )

    # Keep only the columns we’ll need downstream (you can add more later)
    cols = [
        "Driver","Team","LapNumber","Stint","StintLap","Compound","TyreAge",
        "FreshTyre","LapTime_s","DeltaToDriverMedian_s","DeltaToDriverBest_s",
        "TrackTemp","AirTemp","Humidity","WindSpeed","WindDirection","Pressure","Rainfall",
        "Position"
    ]
    return laps[cols].reset_index(drop=True)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

def time_to_seconds(x):
    """Convert 'M:SS.mmm' or 'SS.mmm' to seconds (float)."""
    if pd.isna(x):
        return None
    try:
        if isinstance(x, str) and ":" in x:
            minutes, sec = x.split(":")
            return int(minutes) * 60 + float(sec)
        return float(x)
    except:
        return None

def clean_f1_data(session, year=2023, gp_name="GrandPrix", save_prefix="f1_cleaned"):
    """
    Clean and enrich F1 lap data with weather & track status.
    Saves both raw_cleaned and normalized versions to CSV.
    """
    print(f"🔎 Processing {year} - {gp_name}")

    # Load laps
    laps = session.laps.copy()
    keep_cols = ['Driver', 'DriverNumber', 'LapNumber', 'Stint', 'Compound',
                 'TyreLife', 'FreshTyre', 'LapTime', 'Sector1Time', 'Sector2Time',
                 'Sector3Time', 'Team', 'Position']
    laps = laps[keep_cols]
    
    # Convert Lap/Sector times
    for col in ['LapTime', 'Sector1Time', 'Sector2Time', 'Sector3Time']:
        laps[col] = laps[col].apply(time_to_seconds)
    
    # Add GP info
    laps['GP'] = gp_name

    # --- Weather Data ---
    weather = session.weather_data.copy()
    laps = laps.merge(weather, left_on='LapNumber', right_on='Time', how='left')
    laps = laps.rename(columns={
        'AirTemp': 'AirTemp_C',
        'TrackTemp': 'TrackTemp_C',
        'Humidity': 'Humidity_pct',
        'WindSpeed': 'WindSpeed_mps',
        'WindDirection': 'WindDir_deg',
        'Rainfall': 'Rain'
    }).drop(columns=['Time'])

    # --- Track Status ---
    track_status = session.track_status.copy()
    track_status['Time'] = pd.to_timedelta(track_status['Time'], unit='s')
    laps = laps.merge(track_status, left_on='LapNumber', right_on='Time', how='left')
    laps = laps.rename(columns={'Status': 'TrackStatus'}).drop(columns=['Time'])

    # Handle missing values
    laps = laps.dropna(subset=['LapTime', 'Compound'])
    laps['TyreLife'] = laps['TyreLife'].fillna(0).astype(int)
    laps['FreshTyre'] = laps['FreshTyre'].fillna(False)

    # Remove outliers
    median_lap = laps['LapTime'].median()
    laps = laps[laps['LapTime'] < 1.5 * median_lap]

    # Save non-normalized version
    raw_path = f"{save_prefix}_{year}_{gp_name}_raw.csv"
    laps.to_csv(raw_path, index=False)
    print(f"💾 Saved raw cleaned dataset → {raw_path}")

    # Normalized version
    scaler = MinMaxScaler()
    num_cols = ['LapTime', 'Sector1Time', 'Sector2Time', 'Sector3Time', 'TyreLife',
                'AirTemp_C', 'TrackTemp_C', 'Humidity_pct', 'WindSpeed_mps']
    laps_norm = laps.copy()
    laps_norm[num_cols] = scaler.fit_transform(laps_norm[num_cols])
    norm_path = f"{save_prefix}_{year}_{gp_name}_normalized.csv"
    laps_norm.to_csv(norm_path, index=False)
    print(f"💾 Saved normalized dataset → {norm_path}")

    return laps, laps_norm
```

```python
import fastf1
import pandas as pd
import logging
from fastf1.core import Laps

# Setup logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

def rebuild_2023_with_all_sessions(output_cleaned="cleaned_f1_data_2023.csv",
                                   output_normalized="normalized_f1_data_2023.csv"):
    all_data = []
    races = fastf1.get_event_schedule(2023, include_testing=False)  # Only official season

    for _, event in races.iterrows():
        gp_name = event['EventName']
        logging.info(f"Processing {gp_name} {event['EventDate'].year}")

        # Go through each session: FP1, FP2, FP3, Q, R
        for session_type in ["FP1", "FP2", "FP3", "Q", "R"]:
            try:
                logging.info(f"  Loading {session_type}...")
                session = fastf1.get_session(event['EventDate'].year, event['RoundNumber'], session_type)
                session.load(weather=True)

                # Extract laps
                laps = session.laps
                if laps.empty:
                    logging.warning(f"    No data for {gp_name} {session_type}")
                    continue

                # Add weather info
                weather = session.weather_data
                laps = laps.merge(weather, on="Time", how="left")

                # Add track status if available
                try:
                    track_status = session.get_track_status()
                    laps = laps.merge(track_status, on="Time", how="left")
                except Exception as e:
                    logging.warning(f"    No track status for {gp_name} {session_type} ({e})")

                # Add identifiers
                laps["GP"] = gp_name
                laps["SessionType"] = session_type

                all_data.append(laps)

            except Exception as e:
                logging.error(f"  Failed to process {gp_name} {session_type}: {e}")

    # Combine everything
    if not all_data:
        logging.error("No data extracted!")
        return

    full_df = pd.concat(all_data, ignore_index=True)

    # Select useful columns
    keep_cols = [
        "Driver", "DriverNumber", "LapNumber", "Stint", "Compound", "TyreLife",
        "FreshTyre", "LapTime", "Sector1Time", "Sector2Time", "Sector3Time",
        "Team", "Position", "GP", "SessionType", "Time",
        "AirTemp", "TrackTemp", "Humidity", "Pressure", "Rainfall", "WindSpeed", "WindDirection",
        "Status"  # track status
    ]
    full_df = full_df[[c for c in keep_cols if c in full_df.columns]]

    # Save cleaned (human readable)
    full_df.to_csv(output_cleaned, index=False)
    logging.info(f"Saved cleaned data: {output_cleaned}")

    # Normalized version (LapTime + sectors to seconds)
    normalized_df = full_df.copy()
    for col in ["LapTime", "Sector1Time", "Sector2Time", "Sector3Time"]:
        if col in normalized_df:
            normalized_df[col] = pd.to_timedelta(normalized_df[col]).dt.total_seconds()

    normalized_df.to_csv(output_normalized, index=False)
    logging.info(f"Saved normalized data: {output_normalized}")

    return full_df, normalized_df


# Run pipeline
cleaned, normalized = rebuild_2023_with_all_sessions()
```

```python
import fastf1
import pandas as pd
import logging

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

def extract_gp_data(year: int, gp_name: str, output_prefix="f1_data"):
    """
    Extract FP, Q, and R sessions for one Grand Prix, with weather + track status merged.
    Saves two CSVs: cleaned (readable) and normalized (for ML).
    """
    fastf1.Cache.enable_cache("f1_cache")  # <-- set your cache folder

    session_types = ["FP1", "FP2", "FP3", "Q", "R"]
    all_dfs = []

    for stype in session_types:
        try:
            logging.info(f"Loading {gp_name} {stype}...")
            session = fastf1.get_session(year, gp_name, stype)
            session.load(weather=True)

            laps = session.laps.copy()
            if laps.empty:
                logging.warning(f"No laps found for {gp_name} {stype}, skipping...")
                continue

            # Add GP + session info
            laps["GP"] = gp_name
            laps["SessionType"] = stype

            # Keep the required columns
            keep_cols = [
                "Driver", "DriverNumber", "LapNumber", "Stint", "Compound",
                "TyreLife", "FreshTyre", "LapTime", "Sector1Time", "Sector2Time",
                "Sector3Time", "Team", "Position", "Time", "GP", "SessionType"
            ]
            laps = laps[keep_cols]

            # Merge with weather (nearest timestamp)
            weather = session.weather_data.copy()
            merged = pd.merge_asof(
                laps.sort_values("Time"),
                weather.sort_values("Time"),
                on="Time",
                direction="nearest"
            )

            # Merge with track status (if available)
            if hasattr(session, "track_status"):
                tstatus = session.track_status.copy()
                merged = pd.merge_asof(
                    merged.sort_values("Time"),
                    tstatus.sort_values("Time"),
                    on="Time",
                    direction="nearest"
                )

            all_dfs.append(merged)

        except Exception as e:
            logging.error(f"Failed {gp_name} {stype}: {e}")

    if not all_dfs:
        logging.error(f"No data collected for {gp_name}.")
        return

    full_df = pd.concat(all_dfs, ignore_index=True)

    # Save readable
    cleaned_path = f"{output_prefix}_{year}_{gp_name.replace(' ', '_')}_cleaned.csv"
    full_df.to_csv(cleaned_path, index=False)
    logging.info(f"Saved cleaned data → {cleaned_path}")

    # Save normalized (convert LapTime/Sectors to seconds)
    norm_df = full_df.copy()
    for col in ["LapTime", "Sector1Time", "Sector2Time", "Sector3Time"]:
        norm_df[col] = pd.to_timedelta(norm_df[col]).dt.total_seconds()

    norm_path = f"{output_prefix}_{year}_{gp_name.replace(' ', '_')}_normalized.csv"
    norm_df.to_csv(norm_path, index=False)
    logging.info(f"Saved normalized data → {norm_path}")
extract_gp_data(2023, 'Pre-Season Testing')
```

```python
import fastf1
import pandas as pd
from tqdm import tqdm

# Enable cache
fastf1.Cache.enable_cache('f1_cache')

# Load existing dataset
df_master = pd.read_csv("cleaned_f1_data_2023.csv")

# Missing GPs we need
missing_gps = {
    "Mexico City Grand Prix": 20,
    "São Paulo Grand Prix": 21,
    "Las Vegas Grand Prix": 22,
    "Abu Dhabi Grand Prix": 23
}

year = 2023
all_new_data = []

# Iterate through missing races
for gp_name, round_no in tqdm(missing_gps.items()):
    sessions = ["FP1", "FP2", "FP3", "Q", "R"]  # Free practices, Quali, Race
    
    for session_type in sessions:
        try:
            session = fastf1.get_session(year, round_no, session_type)
            session.load(laps=True, telemetry=False, weather=True)
            
            # Extract laps
            laps = session.laps
            if laps.empty:
                continue
            
            # Add GP + session info
            laps["GP"] = gp_name
            laps["SessionType"] = session_type
            
            # Align schema
            cols = ['Driver', 'DriverNumber', 'LapNumber', 'Stint', 'Compound', 'TyreLife',
                    'FreshTyre', 'LapTime', 'Sector1Time', 'Sector2Time', 'Sector3Time',
                    'Team', 'Position', 'GP', 'SessionType', 'Time', 
                    'AirTemp', 'TrackTemp', 'Humidity', 'Pressure', 'Rainfall', 
                    'WindSpeed', 'WindDirection']
            
            # Add weather columns if available
            if session.weather_data is not None:
                weather = session.weather_data
                # Simplify by merging nearest weather sample on lap start time
                laps = laps.merge(weather, how="left", left_on="Time", right_on="Time")
            else:
                for col in ['AirTemp','TrackTemp','Humidity','Pressure','Rainfall','WindSpeed','WindDirection']:
                    laps[col] = pd.NA
            
            # Keep only required columns (fill missing with NaN)
            for col in cols:
                if col not in laps.columns:
                    laps[col] = pd.NA
            laps = laps[cols]
            
            all_new_data.append(laps)
        
        except Exception as e:
            print(f"Error loading {gp_name} - {session_type}: {e}")

# Combine new races
df_new = pd.concat(all_new_data, ignore_index=True)

# Append to master dataset
df_final = pd.concat([df_master, df_new], ignore_index=True)

# Save
df_final.to_csv("cleaned_f1_2023_updated.csv", index=False)

print("✅ Appended missing races. Saved as cleaned_f1_2023_updated.csv")
```

```python
import pandas as pd

def time_to_seconds(t):
    """
    Convert timedelta or time string into float seconds.
    Handles NaN gracefully.
    """
    if pd.isna(t):
        return None
    
    # If it's already a timedelta
    if isinstance(t, pd.Timedelta):
        return t.total_seconds()
    
    # If it's a string like "0 days 00:01:38.123000"
    try:
        return pd.to_timedelta(t).total_seconds()
    except Exception:
        return None
```

```python
import fastf1
import pandas as pd

# Enable cache (to avoid re-downloading)
fastf1.Cache.enable_cache("f1_cache")

year = 2023
sessions = ["FP1", "FP2", "FP3", "Q", "R", "S"]  # Include Sprint too
weather_records = []

for round_number in range(1, 23):  # 22 races in 2023
    for sess in sessions:
        try:
            session = fastf1.get_session(year, round_number, sess)
            session.load()

            # Weather snapshots (~ every 1 min)
            weather = session.laps.get_weather_data().copy()

            # Add contextual metadata
            weather["GP"] = session.event["EventName"]
            weather["Year"] = year
            weather["SessionType"] = sess
            weather["RoundNumber"] = round_number
            weather["SessionStart"] = session.session_start_time
            weather["SessionEnd"] = session.session_end_time

            weather_records.append(weather)

            print(f"✅ Collected weather for {session.event['EventName']} - {sess}")

        except Exception as e:
            print(f"⚠️ Skipped round {round_number} {sess}: {e}")

# Merge all sessions
weather_df = pd.concat(weather_records, ignore_index=True)

# Sort chronologically: by round number, then session type order
session_order = {"FP1": 1, "FP2": 2, "FP3": 3, "S": 4, "Q": 5, "R": 6}
weather_df["SessionOrder"] = weather_df["SessionType"].map(session_order)

weather_df = weather_df.sort_values(
    by=["RoundNumber", "SessionOrder", "Time"]
).reset_index(drop=True)

# Save to CSV
weather_df.to_csv("weather_2023.csv", index=False)
print(f"💾 Saved → weather_2023.csv ({weather_df.shape[0]} rows)")
```

```python
import fastf1
import pandas as pd
from tqdm import tqdm

fastf1.Cache.enable_cache("f1_cache")  # change folder if needed

def collect_weather(year=2023):
    all_weather = []

    schedule = fastf1.get_event_schedule(year, include_testing=False)
    sessions = ["FP1", "FP2", "FP3", "Q", "R"]

    # total iterations = number of races × number of sessions
    with tqdm(total=len(schedule) * len(sessions), desc=f"Collecting {year} weather") as pbar:
        for rnd, row in schedule.iterrows():
            event = fastf1.get_event(year, rnd)
            for s in sessions:
                try:
                    session = event.get_session(s)
                    session.load()

                    wd = session.weather_data.copy()
                    wd["RoundNumber"] = rnd
                    wd["EventName"] = event.EventName
                    wd["Session"] = s
                    wd["Date"] = event.EventDate

                    all_weather.append(wd)

                except Exception as e:
                    print(f"⚠️ Skipped round {rnd} {s}: {e}")
                finally:
                    pbar.update(1)   # always tick, success or fail

    df_weather = pd.concat(all_weather, ignore_index=True)
    df_weather.sort_values(["RoundNumber", "Session", "Time"], inplace=True)
    df_weather.to_csv(f"weather_{year}.csv", index=False)
    print(f"💾 Saved weather data → weather_{year}.csv")

    return df_weather


# Run once
weather_2023 = collect_weather(2023)
```

```python
# Code to clean and normalise the Weather data (2023)
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

def process_weather(file_in, file_cleaned, file_normalized):
    # Load raw weather data
    raw_weather = pd.read_csv(file_in)

    # ---- Cleaned Weather Data ----
    weather_df = raw_weather.copy()

    # Drop duplicate rows
    weather_df = weather_df.drop_duplicates()

    # Forward fill NaN values
    weather_df = weather_df.fillna(method='ffill')

    # Ensure numeric datatypes
    num_cols = ['AirTemp', 'Humidity', 'Pressure', 'TrackTemp', 'WindDirection', 'WindSpeed']
    weather_df[num_cols] = weather_df[num_cols].astype(float)

    # Save cleaned dataset (with metadata intact)
    weather_df.to_csv(file_cleaned, index=False)
    print(f"💾 Saved cleaned weather data → {file_cleaned}")

    # ---- Normalized Weather Data ----
    norm_weather = weather_df.copy()

    scaler = MinMaxScaler()
    norm_weather[num_cols] = scaler.fit_transform(norm_weather[num_cols])

    # Keep Rainfall as is (binary)
    # Drop metadata for ML
    drop_cols = ['RoundNumber', 'EventName', 'Session', 'Date']
    norm_weather = norm_weather.drop(columns=drop_cols)

    # Save normalized dataset
    norm_weather.to_csv(file_normalized, index=False)
    print(f"💾 Saved normalized weather data → {file_normalized}")


# Example usage
process_weather(
    file_in="weather_2023.csv",
    file_cleaned="weather_2023_cleaned.csv",
    file_normalized="weather_2023_normalized.csv"
)
```
